## Objectifs

- Décomposer un prompt professionnel.
- Comparer des stratégies de classification.
- Évaluer localement les résultats.

## Prompt template

```text
ROLE: Tu es un assistant de triage support.
TASK: Classer un ticket.
OUTPUT: JSON valide.
```

````
from dataclasses import dataclass
from typing import Literal

Category = Literal["BUG", "QUESTION", "FEATURE_REQUEST", "OTHER"]

@dataclass(frozen=True)
class Case:
    ticket: str
    expected: Category
````

````
def classify_ticket(ticket: str) -> Category:
    text = ticket.lower()
    if any(k in text for k in ["erreur", "500", "403", "bug", "plante", "timeout"]):
        return "BUG"
    if any(k in text for k in ["ajouter", "mode sombre", "feature"]):
        return "FEATURE_REQUEST"
    if any(k in text for k in ["comment", "puis-je", "changer", "modifier"]):
        return "QUESTION"
    return "OTHER"
````

````
cases = [
    Case("Le login renvoie une erreur 403 depuis hier.", "BUG"),
    Case("Comment puis-je changer mon adresse e-mail ?", "QUESTION"),
    Case("Pouvez-vous ajouter un mode sombre ?", "FEATURE_REQUEST"),
    Case("Merci pour votre aide.", "OTHER"),
]

for case in cases:
    predicted = classify_ticket(case.ticket)
    print(case.expected, predicted, "PASS" if predicted == case.expected else "FAIL")
````

## Interprétation

Ce notebook illustre la démarche d'évaluation. Dans un système réel, la fonction `classify_ticket` serait remplacée par un appel LLM, puis la sortie serait validée par schéma.
