# Lab — Semantic Search Mock

Objectif : construire une recherche vectorielle simplifiée sans base externe.

```
import numpy as np

 documents = {
    'doc1': 'Politique de congés maladie',
    'doc2': 'Procédure de remboursement des frais',
    'doc3': 'Guide de configuration VPN',
}

vectors = {
    'doc1': np.array([0.9, 0.1, 0.1]),
    'doc2': np.array([0.1, 0.9, 0.2]),
    'doc3': np.array([0.1, 0.2, 0.9]),
}

query = np.array([0.85, 0.15, 0.1])

def cosine(a,b):
    return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))

scores = {doc_id: cosine(query, vector) for doc_id, vector in vectors.items()}
sorted(scores.items(), key=lambda x: x[1], reverse=True)
```
