# Lab NumPy — Embeddings

Objectif : comprendre intuitivement la différence entre one-hot vectors et embeddings denses.

```
import numpy as np

vocab = ['chat', 'chien', 'voiture', 'camion', 'pomme', 'banane']
word_to_id = {w: i for i, w in enumerate(vocab)}
word_to_id
```

```
def one_hot(word):
    v = np.zeros(len(vocab))
    v[word_to_id[word]] = 1
    return v

one_hot('chat')
```

```
chat = one_hot('chat')
chien = one_hot('chien')
voiture = one_hot('voiture')

print('dot(chat, chien)=', np.dot(chat, chien))
print('dot(chat, voiture)=', np.dot(chat, voiture))
```

Dans un one-hot encoding, tous les mots différents sont orthogonaux. Il n'y a pas de proximité sémantique.

```
embeddings = {
    'chat': np.array([0.9, 0.8, 0.1]),
    'chien': np.array([0.85, 0.75, 0.1]),
    'voiture': np.array([0.1, 0.2, 0.9]),
    'camion': np.array([0.1, 0.25, 0.85]),
    'pomme': np.array([0.2, 0.9, 0.2]),
    'banane': np.array([0.25, 0.85, 0.2]),
}

def cosine(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

print('sim(chat, chien)=', cosine(embeddings['chat'], embeddings['chien']))
print('sim(chat, voiture)=', cosine(embeddings['chat'], embeddings['voiture']))
```
